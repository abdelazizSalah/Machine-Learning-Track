### a) uses slice to create a new list made up of only even numbers in the list

```python
import numpy as np

# defining the list 
L = np.arange(10)

#using slicing 
evenArray = L[: :2] # from begining till ending and with step of 2
```

### b) uses slice to create a new list made up of only odd numbers in the list

```python
oddArray = L[ 1: : 2]
```

### c) print both new arrays 

```python
print (evenArray, oddArray, sep='\n-------------------\n')
```

### d) create a list made up from 5 even elements then 5 odd elements 

```python
# OrderedList
L1 = np.arange(0 , 9, 2)
L2 = np.arange(1, 10, 2)
# this is how we concatenate two lists. 
L3 = np.concatenate((L1, L2), axis=0)
print(L3,sep = '\n')
```

### e) delete all even index in the new list

```python
L4 = np.delete(L3, [i % 2 == 0 for i in range (len(L3))])
print(L4)
```

### f) add the element 10 to the list and print it

```python
# append(ListToAppendInto, valueToBeAppended)
L4 = np.append(L4 ,10)
print(L4)
```

### g) print the list after sorting it

```python
print(sorted(L4))

# or 

L4.sort()
print(L4)
```

### h) print the list in reverse order

```python
print(np.flip(L4))
```

### i) create a new list made up of the first element from the last list repeated 5 times

```python
L5 = np.repeat(L4[0], 5)
print(L5)
```
