### Cleaning the data
* we have to clean the data, cause it's not always clean.
* so we need to remove all the null data. 
* so first we have to check if there is any null data.
* df.isnull() -> this gets you all the null data

### Dealing with null data
1. we drop the column that has null data 
   1. we do this if the percentage of null in the column is relativly high.
   2. to drop a column we use df.drop(columns = ['col_name1', 'col_name2', ...] ,inplace=True)
2. we may fill the null data, to make use of the data we have
   1. Steps:
      1. determine the data type of the column, whether it's numeric or categorical
      2. if it was catergorical, we fill it with the most frequent value in the column (mode), because this make sense. 
      3. if it was numeric, we fill it with the (mean, mode, median or statistical data) of the column.
      4. we need to use mode with numeric data, if we have a survey. 
      5. mean is the average of the data.
      6. median is the element in the middle.

```python
# this returns all the data that has a null value
df.isnull()

# to get the percentage of null values in each column
print(df.isnull().mean() * 100)

# if this percentage is high, we can drop the column
df.drop(columns = ['region_2'], inplace = True)
```

```python
x = np.array([1,2,5,5,5])

## or we can get it from pandas
x = pd.Series(x)
x.mode() # the most frequent value
x.median() # the middle value
x.mean() # the average
```

* when we call df['col_name'] this return a series, so to be able to act on we have to acces the first element. 
* df['col_name'][0] -> this will return the first element in the column.
* if we have two elements with the same frequency, .mode() will return the all of them.

```python
# fillna -> fill the null values
# we fill it with values 
df['newCountry'].fillna(df['newCountry'].mode()[0], inplace = True) 
df
```
